```python
dict_list = []
for col in df.columns:
    col_nulls = df[col].isnull().sum()
    col_null_percent = col_nulls/len(df.index)
    dict_list.append({'':col,'num_rows_missing':col_nulls,'pct_rows_missing':col_null_percent})
pd.DataFrame(dict_list).set_index('').sort_values(by='num_rows_missing', ascending=False).head(34)
```

```python
prop_required_column = .75
prop_required_row = .75
```

```python
threshold = int(round(prop_required_column*len(df.index),0))
df.dropna(axis=1, thresh=threshold, inplace=True)
threshold = int(round(prop_required_row*len(df.columns),0))
df.dropna(axis=0, thresh=threshold, inplace=True)
df.info()
```
